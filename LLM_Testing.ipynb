```python

from typing import Tuple
import random

class Agent(object):
    def __init__(self, name, position, color, tool_property, inventory, status="idle"):
        self._name = name
        self._position = position
        self._color = color
        self._goal = None
        self._status = status
        self.tool_property = tool_property
        self._inventory = []
        

    def setGoal(self, goal):
        """Sets the Goal for the robot"""
        self._goal = goal
        self.set_status("active")


    def set_status(self, status="idle"):
        if status in ["idle", "active", "charging"]:
            old_status = self._status
            self._status = status
            if self._status == "idle":
                self._goal = None
    
    def add_item_invetory(self, item):
        self._inventory.append(item)

    @property
    def tool_property(self) -> str:
        """agent tool i.e gripper"""
        return self._tool_property
    
    @tool_property.setter
    def tool_property(self, new_tool):
        self._tool_property = new_tool

    @property
    def inventory(self) -> list:
        """ An agents inventory """
        return self._inventory

    @property
    def position(self) -> Tuple[int, int]:
        """ The position of the agent """
        return self._position

    @position.setter
    def position(self, new_position: Tuple[int, int]):
        """ Set the new position

        :new_position: The new node id

        """
        self._position = new_position

    @property
    def name(self) -> str:
        """ The name of the agent """
        return self._name

    @name.setter
    def name(self, new_name: str):
        """ Sets a new name for the agent

        :new_name: The new name for the agent

        """
        self._name = new_name

    @property
    def color(self) -> str:
        """ The color of the agent """
        return self._color

    @color.setter
    def color(self, new_color: str):
        """ Sets a new color for the agent

        :new_goal: The new color for the agent

        """
        self._color = new_color

    @property
    def goal(self) -> Tuple[int, int]:
        """ The goal of the agent """
        return self._goal

    @goal.setter
    def goal(self, new_goal: Tuple[int, int]):
        """ Set the new goal

        :new_goal: The new goal node id

        """
        self._goal = new_goal
```

## Create Tools

**Helper functions**

```python
def sort_by_tool(tool_name, robot_list=robots):
    usable_robots = []
    for robot in robot_list:
        if robot.tool_property == tool_name:
            usable_robots.append(robot)
    return usable_robots

def sort_by_availability(robot_list, target_pos):
    available_robots = []
    if robot_list:
        closest_robot = robot_list[0].name
        min_dist = dist_func(robot_list[0].position, target_pos)
        flag = True
        for robot in robot_list:
            if robot._status != "active":
                available_robots.append(robot.name)
                if flag:
                    flag = False
                    closest_robot = robot.name
                    min_dist = dist_func(robot.position, target_pos)
                    
                if min_dist > dist_func(robot.position, target_pos):
                    closest_robot = robot.name
                    min_dist = dist_func(robot.position, target_pos)
        if available_robots:
            return f"out of available robot: {available_robots}, {closest_robot} is the best suited to solve the task."
    return "no robot with the specified tool to solve the task"

def dist_func(robot_pos, target_pos):
    x_dist = abs(target_pos["x"] - robot_pos[0])
    y_dist = abs(target_pos["y"] - robot_pos[1])
    total_dist = x_dist + y_dist
    return total_dist

def select_known_robot(robot_name, desired_tool, robot_list=robots):
    for robot in robot_list:                
        if robot_name == robot.name:
            if robot._status != "active":
                if desired_tool == robot.tool_property:
                    return f"{robot_name} with the tool property: {robot.tool_property} has been selected to solve the task"
                return f"""{robot_name} is not equipped with the tool property initially selected for task completion. 
                Can {robot_name} with tool property: {robot.tool_property} still solve the task?
                In your anwser please stat the initial selected tool and why {robot_name} can/can't solve the task with the tool property: {robot.tool_property}. 
                """
            return f"{robot_name} is currently active and is not available to solve the task"
    return f"{robot_name} is not a valid robot"

def change_status(robot_name, robot_list=robots, status="idle"):
    for robot in robots:
        if robot_name == robot.name:
            old_status = robot._status
            robot.set_status(status)
            return f"{robot.name} has changed status from {old_status} to {status}"

def set_goal(robot_name, position, robot_list=robots):
    for robot in robots:
        if robot_name == robot.name:
            robot.setGoal(position)
            return robot.goal
        
def add_item_to_inventory(robot_name, item_name, robot_list=robots):
    for robot in robots:
        if robot_name == robot.name:
            robot.add_item_invetory(item_name)
```

**LLM tools**

```python
from typing import Annotated, List, Tuple, Union

from langchain_core.tools import tool
from langsmith import trace

@tool
def move_robot(robot_name: str, position: Tuple) -> str:
    """Moves the Specified robot to the desired position"""
    goal = set_goal(robot_name=robot_name, position=position)
    return f"{robot_name} has been moved to the desired position {goal}"

@tool
def grab_or_remove_item(robot_name: str, item_name: str) -> str:
    """makes the specified robot grab/remove the desired item"""
    add_item_to_inventory(robot_name=robot_name, item_name=item_name)
    return f"{robot_name} has grabed/removed the desired item {item_name}"

@tool
def robot_status_change(robot_name: str, status: str):
    """Changes the status of a robot"""
    return change_status(robot_name=robot_name, status=status)

@tool
def robot_selection_tool(desired_robot_tool_property, 
                         item_position) -> str:
    """Finds the most robots with the desired tool property. Remember that the item_position needs to be a dict of x and y coordinates"""
    sorted_robots = sort_by_tool(desired_robot_tool_property)
    return sort_by_availability(sorted_robots, item_position)
@tool
def select_known_robot_tool(robot_name: str, desired_robot_tool_property) -> str:
    """select the specified robot to solve the task. If the robot does not have the desired tool for the task, then it will ask if the robot still can solve the task.
    This tool does not alter the tools a robot has, but will try to use the selected robot although it might not have the desired tool.
    """
    return select_known_robot(robot_name=robot_name, desired_tool=desired_robot_tool_property)

@tool
def agent_response_tool(input: str, response: str) -> str:
    """this is a tool that returns the response to the input"""
    return response
```

## Helper Utilities

We are going to create a few utility functions to make it more concise when we want to:

1. Create a worker agent.
2. Create a supervisor for the sub-graph.

These will simplify the graph compositional code at the end for us so it's easier to see what's going on.

```python
from typing import Any, Callable, List, Optional, TypedDict, Union

from langchain.agents import AgentExecutor, create_openai_functions_agent, create_tool_calling_agent
from langchain.output_parsers.openai_functions import JsonOutputFunctionsParser
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables import Runnable
from langchain_core.tools import BaseTool
from langchain_openai import ChatOpenAI

from langgraph.graph import END, StateGraph


def create_agent(
    llm: ChatOpenAI,
    tools: list,
    system_prompt: str,
) -> str:
    """Create a function-calling agent and add it to the graph."""
    system_prompt += """\nWork autonomously according to your specialty, using the tools available to you.
    Do not ask for clarification.

    Important!: When giving the final output to the supervisor, give a short summary of what your thought process was when selecting your tools an also the results. Do the summary in bulletpoint format.
    
    EXAMPLE
    Thought process
    - insert process here

    Results:
    - insert results here
    END OF EXAMPLE

    Your other team members (and other teams) will collaborate with you with their own specialties.
    You are chosen for a reason!
    """
    prompt = ChatPromptTemplate.from_messages(
        [
            (
                "system",
                system_prompt,
            ),
            MessagesPlaceholder(variable_name="messages"),
            MessagesPlaceholder(variable_name="agent_scratchpad"),
        ]
    )
    agent = create_tool_calling_agent(llm, tools, prompt)
    executor = AgentExecutor(agent=agent, tools=tools)
    return executor


def agent_node(state, agent, name):
    result = agent.invoke(state)
    return {"messages": [HumanMessage(content=result["output"], name=name)]}


def create_team_supervisor(llm: ChatOpenAI, system_prompt, members) -> str:
    """An LLM-based router."""
    options = ["FINISH"] + members
    function_def = {
        "name": "route",
        "description": "Select the next role.",
        "parameters": {
            "title": "routeSchema",
            "type": "object",
            "properties": {
                "next": {
                    "title": "Next",
                    "anyOf": [
                        {"enum": options},
                    ],
                },
                "response": {
                    "type": "string",
                    "description": "The reason behind the choice of member or a description of the task completion to the user."
                }
            },
            "required": ["response", "next"],
        },
    }
    prompt = ChatPromptTemplate.from_messages(
        [
            ("system", system_prompt),
            MessagesPlaceholder(variable_name="messages"),
            (
                "system",
                """Given the conversation above, who should act next? Or should we FINISH? Select one of: {options}. 
                If we should FINISH then decribe the task completion in the following format:
                
                EXAMPLE
                ROBOT NAME;TOOL PROPERTY;ITEM;ITEM POSITION
                EXAMPLE END
                
                ROBOT NAME = (The robot that solved the task i.e Robot22)
                TOOL PROPERTY = (The tool property of ROBOT i.e gripper)
                ITEM = (The item that is being manipulated i.e toolkit)
                ITEM POSITION = (The position of ITEM i.e (11, 8))
                
                Remember to respond exactly as in the example but be mindful of the upper and lower case for the name of the robot, tool property and item. If we should not FINISH then respond with the reasoning behind the choice."""
            ),
        ]
    ).partial(options=str(options), team_members=", ".join(members))
    return (
        prompt
        | llm.bind_functions(functions=[function_def], function_call="route")
        | JsonOutputFunctionsParser()
    )
```

## Define Agent Teams

**Robot Control Team**

The team will have the ability to control the robots directly using function calling

```python
import functools
import operator

from langchain_core.messages import AIMessage, BaseMessage, HumanMessage
from langchain_openai.chat_models import ChatOpenAI
import functools

# Robot Control Team graph state
class RobotControlTeamState(TypedDict):
    # A message is added after each team member finishes
    messages: Annotated[List[BaseMessage], operator.add]
    # The team members are tracked so they are aware of
    # the others' skill-sets
    team_members: List[str]
    # Used to route work. The supervisor calls a function
    # that will update this every time it makes a decision
    next: str
    # The reason behind the member choice
    response: str

# general llm used in all agents
llm = ChatOpenAI(model="gpt-4-1106-preview")

# agent used to control the actions of the robots
robot_control_agent = create_agent(
    llm,
    [move_robot, grab_or_remove_item, robot_status_change],
    """
    You are an agent that control robots actions using the tools available to you. When the robot is done finishing the task it should always have it's status set to idle.
    remember you are only able to grab/remove the items and not the robots tool property. this is the list of tool properties: [lawnmower, gripper, tow hook, saw, welding tool, mop, spray nozzle, broom, front loader, vacuum].
     which are not to be grabed or removed."""
)

robot_control_node = functools.partial(agent_node, agent=robot_control_agent, name="Controller")

# agent to determine the best suited robot to solve the task
robot_selection_agent = create_agent(
    llm,
    [robot_selection_tool, select_known_robot_tool],
    f"""
    You are a agent with the purpose to select the best suited robot for the task. A robot can have the following tool properties: [lawnmower, gripper, tow hook, saw, welding tool, mop, spray nozzle, broom, front loader, vacuum].
    If a specified robot is already selected to solve the task, use select_known_robot_tool.
    Here is an example of that.

    EXAMPLE: 
    task = "Move robot_name to location (2, 11) and sort the recyclables."
    Robot = robot_name
    END EXAMPLE

    remember to select the most apropriate tool_property for the task i.e a gripper for a task where something needs to be picked up.
     
"""
)

robot_selection_node = functools.partial(agent_node, agent=robot_selection_agent, name="Robot_selector")

# agent that answer general question such as "what is the capital of denmark"
misc_agent=create_agent(
    llm,
    [agent_response_tool],
    """
    You are an agent that can answer general questions and messages that doesn't have anything to do with making a robot solve a complex task. When you know the apropriate reponse return the answer to the supervisor.
    """
)

misc_node=functools.partial(agent_node, agent=misc_agent, name="misc_agent")


# This is the supervisor agent tasked with selecting the next worker (agent). the supervisor might select multiple workers before the task given is consitered solved.
supervisor_agent = create_team_supervisor(
    llm,
    """
    You are a supervisor tasked with managing a conversation between the following workers: Controller, Robot_selector, misc_agent.
    Given the following user request, respond with the worker to act next and a response depending if we are FINISH or not.
    Each worker will perform a task and respond with their results and status. If the response from the Robot_selector determines that the robot is not suitable to solve the task, then either finish or make the Robot_selector
    When finished, respond with FINISH.
    """,
    ["Controller", "Robot_selector", "misc_agent"],
)
```

**Connect the nodes in the graph**

```python
#adding the nodes to the graph
robot_control_graph = StateGraph(RobotControlTeamState)
robot_control_graph.add_node("Controller", robot_control_node)
robot_control_graph.add_node("Robot_selector", robot_selection_node)
robot_control_graph.add_node("misc_agent", misc_node)
robot_control_graph.add_node("supervisor", supervisor_agent)

#define the control flow
robot_control_graph.add_edge("Controller", "supervisor")
robot_control_graph.add_edge("Robot_selector", "supervisor")
robot_control_graph.add_edge("misc_agent", "supervisor")
robot_control_graph.add_conditional_edges(
    "supervisor",
    lambda x: x["next"],
    {"Controller": "Controller", "Robot_selector": "Robot_selector", "misc_agent": "misc_agent", "FINISH": END}
)

robot_control_graph.set_entry_point("supervisor")
chain = robot_control_graph.compile()


# The following functions interoperate between the top level graph state
# and the state of the robot control sub-graph
# this makes it so that the states of each graph don't get intermixed
def enter_chain(message: str):
    results = {
        "messages": [HumanMessage(content=message)],
    }
    return results

robot_control_chain = enter_chain | chain
```

```python
def task_iteration(file_path):
        task = file_path.readline()
        if not task:
            return False
        return task.strip()
```

```python
def clear_txt(path):
    # Clear the file by opening it in write mode and closing it
    with open(path, "w") as file:
        pass

def write_txt(path, string):
    # Open the file in append mode and write the string
    with open(path, "a") as file:
        file.write(string)

def process_dict(d, path):
    output = []

    if 'supervisor' in d:
        supervisor_info = d['supervisor']
        next_step = supervisor_info.get('next', 'N/A')
        response = supervisor_info.get('response', 'N/A')
        if next_step == "FINISH":
            write_txt(path, f"{response}\n")
        output.append(f"Supervisor:\nNext: {next_step}\nresponse: {response}\n---\n")
        print(f"Supervisor:\nNext: {next_step}\nresponse: {response}\n---\n")
        
    if 'Robot_selector' in d:
        robot_selector_info = d.get('Robot_selector', {})
        messages = robot_selector_info.get('messages', [])
        for message in messages:
            content = getattr(message, 'content', 'N/A')
            output.append(f"Robot Selector:\n{content}\n---\n")
            print(f"Robot Selector:\n{content}\n---\n")

    if 'Controller' in d:
        controller_info = d.get('Controller', {})
        messages = controller_info.get('messages', [])
        for message in messages:
            content = getattr(message, 'content', 'N/A')
            output.append(f"Controller:\n{content}\n---\n")
            print(f"Controller:\n{content}\n---\n")
    
    if 'misc_agent' in d:
        misc_agent_info = d.get('misc_agent', {})
        messages = misc_agent_info.get('messages', [])
        for message in messages:
            content = getattr(message, 'content', 'N/A')
            output.append(f"misc_agent:\n{content}\n---\n")
            print(f"misc_agent:\n{content}\n---\n")
    

    return ''.join(output)
```

```python
import csv
file_path = "Easy_matrix.txt"
clear_txt(file_path)
# Open the CSV file once outside the loop for appending data
with open('output_Easy.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=";")

    # Open the task file and read tasks one by one
    with open("Easy_tasks_test.txt", 'r') as file:
        i = 1
        while True:
            history = ""
            task = task_iteration(file)
            if not task:
                break
            
            # Print the task number and task to the console
            print(f"{i}. task: {task}")
            print("---")

            # Process each stream associated with the task
            for s in robot_control_chain.stream(
                task, {"recursion_limit": 100}
            ):
                if "__end__" not in s:
                    history += process_dict(s, file_path)
            i += 1
            
            # Append both task and processed data to the CSV
            writer.writerow([task, history])
        
```
